```python
def generate_blowfish_key(key_length_bits):
    if key_length_bits < 32 or key_length_bits > 448:
        raise ValueError("Key length must be between 32 and 448 bits")

    # Generate a random key of the specified length
    key_bytes = bytearray((key_length_bits + 7) // 8)
    key = 0

    for i in range(len(key_bytes)):
        key_bytes[i] = ord(os.urandom(1)) 

    for i in range(len(key_bytes)):
        key = (key << 8) | key_bytes[i]

    return key

if __name__ == "__main__":
    key_length = 128  
    key = generate_blowfish_key(key_length)
    print(f"Generated Blowfish key of {key_length} bits: {key:0{key_length//4}x}")
```

```python
mod_val = 1  # 2^32 or 2^64 depending on the block size

def hex_to_bin(hex_string):
    binary = ""
    n = len(hex_string)
    for i in range(n):
        num = int(hex_string[i], 16)
        binary_4b = bin(num)[2:]
        
        # Ensure each value in hexadecimal is 4 bits in binary.
        binary_4b = "0000" + binary_4b
        binary_4b = binary_4b[-4:]  # Keep the last 4 bits
        
        binary += binary_4b
    
    return binary

#add print statements to see what is going on
hex_string = "3A5F"
binary = hex_to_bin(hex_string)
print(f"Hexadecimal string: {hex_string}")
print(f"Binary string: {binary}")
```

```python
def bin_to_hex(binary_string):
    hex_string = ""
    n = len(binary_string)
    # Ensure the length of binary_string is a multiple of 4 by padding with zeros
    padding = (4 - (n % 4)) % 4
    binary_string = "0" * padding + binary_string
    for i in range(0, n, 4):
        binary_4b = binary_string[i:i + 4]
        num = int(binary_4b, 2)
        hex_string += format(num, 'X')
    return hex_string

#add print statements to see what is going on
binary_string = "0011101001011111"
hex_string = bin_to_hex(binary_string)
print(f"Binary string: {binary_string}")
print(f"Hexadecimal string: {hex_string}")
```

```python
def xor_hex(hex_string1, hex_string2):
    # Convert hexadecimal strings to binary
    binary_string1 = hex_to_bin(hex_string1)
    binary_string2 = hex_to_bin(hex_string2)

    # Ensure both binary strings have the same length
    max_len = max(len(binary_string1), len(binary_string2))
    binary_string1 = binary_string1.zfill(max_len)
    binary_string2 = binary_string2.zfill(max_len)

    binary_string = ""
    for i in range(max_len):
        if binary_string1[i] == binary_string2[i]:
            binary_string += "0"
        else:
            binary_string += "1"

    return bin_to_hex(binary_string)

#add print statements to see what is going on
hex_string1 = "3A5F"
hex_string2 = "6D4B"
hex_string = xor_hex(hex_string1, hex_string2)
print(f"Hexadecimal string 1: {hex_string1}")
print(f"Hexadecimal string 2: {hex_string2}")
print(f"XOR of hexadecimal strings: {hex_string}")
```

```python
#add two hexadecimal strings for 64 bit input
def add_hex(hex_string1, hex_string2):

    # Convert hexadecimal strings to binary
    binary_string1 = hex_to_bin(hex_string1)
    binary_string2 = hex_to_bin(hex_string2)

    # Ensure both binary strings have the same length
    max_len = max(len(binary_string1), len(binary_string2))
    binary_string1 = binary_string1.zfill(max_len)
    binary_string2 = binary_string2.zfill(max_len)

    binary_string = ""
    carry = 0
    for i in range(max_len - 1, -1, -1):
        sum = int(binary_string1[i]) + int(binary_string2[i]) + carry
        if sum == 0:
            binary_string += "0"
            carry = 0
        elif sum == 1:
            binary_string += "1"
            carry = 0
        elif sum == 2:
            binary_string += "0"
            carry = 1
        elif sum == 3:
            binary_string += "1"
            carry = 1

    binary_string = binary_string[::-1]
    return bin_to_hex(binary_string)
    
#add print statements to see what is going on
hex_string1 = "FEDCBA9876543210" #
hex_string2 = "123456789ABCDEF0"
hex_string = add_hex(hex_string1, hex_string2)
print(f"Hexadecimal string 1: {hex_string1}")
print(f"Hexadecimal string 2: {hex_string2}")
print(f"Addition of hexadecimal strings: {hex_string}")
```

```python
def round_function(time, plain_text, P_ARRAY): 
    left = plain_text[:8]
    right = plain_text[8:16]

    left = xor_hex(left, P_ARRAY[time])

    # Output from F function
    f_out = f(left)

    right = xor_hex(right, f_out)

    print(f"round {time}: {right}{left}")

    # Swap left and right
    return right + left
```

```python
def encrypt(plain_text, P_ARRAY):
    for i in range(16):
        plain_text = round_function(i, plain_text, P_ARRAY)

    # Postprocessing
    right = plain_text[:8]
    left = plain_text[8:16]

    right = xor_hex(right, P_ARRAY[16])
    left = xor_hex(left, P_ARRAY[17])

    return left + right
```
